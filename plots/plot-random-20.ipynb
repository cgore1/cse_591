```python
#citeseerx
# 20

x = [0.0001,0.001,0.01,0.1]
#with plt.style.context('fivethirtyeight'):
#20%
y = [1441,1953,6962,54909]
plt.plot(x,y, label='80% SpaReach', linewidth=1, color='red', marker = '^')


#40
x = [0.0001,0.001,0.01,0.1]
y = [240,229,154,64]
plt.plot(x,y, label='80% Traversal', linewidth=1, color='green', marker = 'o')



plt.axis([min(x)/2, max(x) * 2, min(y)/2, max(y) * 2000])
plt.legend(loc='upper left')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Query Time (ms)')
plt.xlabel('Spatial Range')
plt.tight_layout()
plt.savefig('C:\\Users\\cHinmaY\\cse_591\\plot1.eps', format='eps', transparent='true', facecolor='white', dpi=1000)
plt.savefig('C:\\Users\\cHinmaY\\cse_591\\plot1.jpeg', format='jpeg', transparent='true', facecolor='white', dpi=1000)
plt.show()
```
