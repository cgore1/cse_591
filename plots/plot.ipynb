```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
%matplotlib inline
```

```python
#citeseerx

x = [0.0001,0.001,0.01,0.1,1]
with plt.style.context('fivethirtyeight'):
    #20%
    y = [1441,1953,6962,54909,323874]
    plt.plot(x,y, label='80%', linewidth=1, color='red', marker = '^')


    #40
    x = [0.0001,0.001,0.01,0.1]
    y = [1333,1706,5411,41173]
    plt.plot(x,y, label='60%', linewidth=1, color='darkcyan', marker = 'o')


    #60
    x = [0.0001,0.001,0.01,0.1]
    y = [1228,1480,3797,26930]
    plt.plot(x,y, label='40%', linewidth=1 , color='green', marker = 'v')
    
    #80
    x = [0.0001,0.001,0.01,0.1]
    y = [1070,1330,2335,13934]
    plt.plot(x,y, label='20%', linewidth=1 , color='blue', marker = 'x')
   


    plt.axis([min(x)/2, max(x) * 10 + 1, min(y)/10, max(y) * 30])
    plt.legend(loc='upper left', shadow=True)
    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel('Query Time (ms)')
    plt.xlabel('Spatial Range')
    plt.show()
```
